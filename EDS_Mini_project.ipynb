```python
# Import all required libraries and read the dataset containing the IIT admissions data.
import csv
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
f=pd.read_csv("student_data.csv")
f
```

```python
# Basic Data Analysis - Calculating records
num_records = len(f)
print("Number of records before cleaning :", num_records)
```

```python
# Dropping rows with unavailable values
f = f.dropna()
f
```

```python
# Cleaning the dataset by applying certain conditions
df=f[f['Expected Year of Graduation']-f['Year of Admission']>=4]
df
```

```python
# Calculate the average fees and discount
avg_fees = df['Fees'].mean()
avg_discount = df['Discount on Fees'].mean()
print("Average fees:", avg_fees)
print("Average discount on fees:", avg_discount)
```

```python
# Calculate the average fees and discount based on field of study
avg_fees_field = df.groupby('Field of Study')['Fees'].mean()
avg_disc_field = df.groupby('Field of Study')['Discount on Fees'].mean()
print(avg_fees_field,'\n')
print(avg_disc_field)
```

```python
max_fees_field = df.groupby('Field of Study')['Fees'].max()
min_fees_field = df.groupby('Field of Study')['Fees'].min()
print(max_fees_field,'\n')
print(min_fees_field)
```

```python
# Calculate the average fees and discount based on year of admission
avg_fees_year = df.groupby('Year of Admission')['Fees'].mean()
avg_disc_year = df.groupby('Year of Admission')['Discount on Fees'].mean()
print(avg_fees_year,'\n')
print(avg_disc_year)
```

```python
max_fees = df.groupby('Year of Admission')['Fees'].max()
print(max_fees,'\n')
min_fees = df.groupby('Year of Admission')['Fees'].min()
print(min_fees)
```

```python
# Determine the most common field of study and specialization
most_common_field_of_study = df['Field of Study'].mode().values[0]
most_common_specialization = df['Specialization'].mode().values[0]

print('Most common field of study:',most_common_field_of_study)
print('Most common specialization:',most_common_specialization)
```

```python
# Count the number of students in each field of study
field_counts = df['Field of Study'].value_counts()
print(field_counts)
```

```python
# Analyze the distribution of expected years of graduation
expdyrofgrn = df['Expected Year of Graduation'].value_counts()
expdyrofgrn
```

```python
dobs = df['Date of Birth'].tolist()
print(dobs)
```

```python
from datetime import datetime

# Empty list to store the extracted years
years = []

# Iterate over each date
for date_str in dobs:
    # Convert the string to a datetime object
    date_obj = datetime.strptime(date_str, '%d-%m-%Y')

    # Extract the year from the datetime object
    year = date_obj.year

    # Append the year to the list
    years.append(year)

# Print the extracted years
print(years)
```

```python
ymax=max(years)
print(ymax)
ymin=min(years)
print(ymin)
ymean=sum(years)/len(years)
print(ymean)
```

```python
# Step 5: Generate visualizations
df['Year of Admission'].hist()
plt.xlabel('Year of Admission')
plt.ylabel('Total Admissions')
plt.title('Distribution of Year of Admission')
plt.show()
```

```python
# Load the dataset
data = pd.read_csv('student_data.csv')

# Distribution of Students by Field of Study
field_counts = data['Field of Study'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(field_counts.index, field_counts.values)
plt.xlabel('Field of Study')
plt.ylabel('Number of Students')
plt.title('Distribution of Students by Field of Study')
plt.xticks(rotation=45)
plt.show()

# Distribution of Students by Current Semester
semester_counts = data['Current Semester'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(semester_counts.index, semester_counts.values)
plt.xlabel('Current Semester')
plt.ylabel('Number of Students')
plt.title('Distribution of Students by Current Semester')
plt.show()

# Specialization of Students
specialization_counts = data['Specialization'].value_counts()

plt.figure(figsize=(10, 6))
plt.pie(specialization_counts.values, labels=specialization_counts.index, autopct='%1.1f%%')
plt.title('Specialization of Students')
plt.axis('equal')
plt.show()

# Fees and Discount on Fees
fees = data['Fees']
discounts = data['Discount on Fees']

plt.figure(figsize=(8, 6))
plt.scatter(fees, discounts)
plt.xlabel('Fees')
plt.ylabel('Discount on Fees')
plt.title('Fees vs. Discount on Fees')
plt.show()
```
